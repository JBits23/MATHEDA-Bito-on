```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\romme\Desktop\MATHEDA Bito-on\WK7\raw\chip-dataset-cleaned.csv",
                 delimiter=",")
df.info()
```

# N Chips Manufactured

```python
df_w = df.copy()
```

```python
chip_count = df_w['Vendor'].value_counts()
chip_count
```

```python
type(chip_count)
```

```python
chart = chip_count.plot(kind='bar', color = ['red', 'orange', 'yellow', 'green', 'blue'], figsize = (6,6) )
chart.bar_label(chart.containers[0], label_type='edge')
plt.title('Chips Manufactured (CPU+GPU)', fontfamily = 'Times New Roman', fontsize = 16)
plt.ylabel('Number of Chips' , fontfamily = 'Times New Roman', fontsize = 16)
plt.xlabel('Manufacturer' , fontfamily = 'Times New Roman', fontsize = 16)
```

# Lowest TDP (Thermal Design Power)

```python
df_tdp = df_w[['Product','TDP']]
df_tdp
```

```python
type(df_tdp)
```

```python
df_tdp = df_tdp.sort_values(by='TDP')
df_tdp
```

# GPU Process Size Trend

```python
df_ps = df_w[['Type','Date','Process_Size']].where(df_w['Type']=='GPU').dropna()
df_ps
```

```python
df_ps['Date'] = pd.to_datetime(df_ps['Date'])
df_ps['Year'] = df_ps['Date'].dt.year
df_ps 
```

```python
df_ps = df_ps.groupby('Year')['Process_Size'].mean().reset_index()
df_ps
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_ps['Year'],df_ps['Process_Size'], color= 'green')
plt.title('GPU Process Size Trend' , fontfamily = 'Times New Roman', fontsize = 16)
plt.ylabel('Transistor Size (nm)' , fontfamily = 'Times New Roman', fontsize = 16)
```
